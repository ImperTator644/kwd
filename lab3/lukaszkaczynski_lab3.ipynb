```python
import pandas as pd
import numpy as np

from pandas import ExcelWriter
from pandas import ExcelFile
```

### Loading data

```python
credit_clients_file = pd.read_excel('credit_clients.xls')
credit_clients_file.head()
```

```python
class DataSet:
    def __init__(self, data = None, target = None):
        self.data = data
        self.target = target
```

```python
clients_data = credit_clients_file.iloc[1:,1:-1]
clients_target = credit_clients_file.iloc[1:,-1]

clients_data.head(20)
```

```python
credit_clients = DataSet(np.array(clients_data, dtype=np.int32), np.array(clients_target, dtype=np.int32))

print(type(credit_clients.data))
print(type(credit_clients.target))

print(credit_clients.data.shape)
print(credit_clients.target.shape)
```

```python
print("First customer in database before scaling")
print(credit_clients.data[0])
print("Should have credit?")
print(credit_clients.target[0])
```

```python
print("---Mean---")
print(credit_clients.data.mean(axis=0))
print("---std---")
print(credit_clients.data.std(axis=0))
```

## Scaling data with StandardScaler

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled_credit_clients = DataSet(scaler.fit_transform(credit_clients.data), credit_clients.target)
```

```python
print("First client in database after scaling")
print(scaled_credit_clients.data[1])
print('---Mean---')
print(scaled_credit_clients.data.mean(axis=0))
print('--std---')
print(scaled_credit_clients.data.std(axis=0))
```

### Split data in train and test batches

```python
from sklearn.model_selection import train_test_split

clients_train_set = DataSet()
clients_test_set = DataSet()

clients_train_set.data, clients_test_set.data, \
clients_train_set.target, clients_test_set.target = \
train_test_split(scaled_credit_clients.data, scaled_credit_clients.target, test_size=0.1)
```

```python
print("Training dataset:")
print("clients_train_set.data:", clients_train_set.data.shape)
print("clients_train_set.target:", clients_train_set.target.shape)
```

```python
print("Testing dataset:")
print("clients_test_set.data:", clients_test_set.data.shape)
print("clients_test_set.target:", clients_test_set.target.shape)
```

### Initiate and train model on training data

```python
from sklearn.linear_model import LogisticRegression

credit_clients_logistic = LogisticRegression()
credit_clients_logistic.fit(clients_train_set.data, clients_train_set.target)
```

### Using classifier

```python
id = 2
credit_prediction = credit_clients_logistic.predict(clients_test_set.data[id,:].reshape(1,-1))
print("Model predicted for client {0} value {1}".format(id+1, credit_prediction))

print("Real value for client \"{0}\" is {1}".format(id+1, clients_test_set.target[id]))
```

```python
credit_prediction_probability = credit_clients_logistic.predict_proba(clients_test_set.data[id,:].reshape(1,-1))
print(credit_prediction_probability)
```

### Evaluate a classifier using the test data

```python
from sklearn.metrics import accuracy_score

credit_acc = accuracy_score(clients_test_set.target, credit_clients_logistic.predict(clients_test_set.data))
print("Model accuracy is {0:0.5f}".format(credit_acc))
```

### Evaluating classifier using confusion matrix

```python
from sklearn.metrics import confusion_matrix

credit_conf_matrix = confusion_matrix(clients_test_set.target, credit_clients_logistic.predict(clients_test_set.data))
print(credit_conf_matrix)
```
